```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('Diabetes.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1:].values
xtrain, xtest, ytrain, ytest = tts(x, y, test_size = 0.25, random_state=0)
sc = StandardScaler()
xtrain[:, [0,1,2,3,4,5,7]] = sc.fit_transform(xtrain[:, [0,1,2,3,4,5,7]])
xtest[:, [0,1,2,3,4,5,7]] = sc.transform(xtest[:, [0,1,2,3,4,5,7]])
```

```python
c = LogisticRegression()
c.fit(xtrain, ytrain)
ypred = c.predict(xtest)
print(accuracy_score(ytest, ypred))
```

```python
c = DecisionTreeClassifier()
c.fit(xtrain, ytrain)
ypred = c.predict(xtest)
print(accuracy_score(ytest, ypred))
```

```python
c = RandomForestClassifier()
c.fit(xtrain, ytrain)
ypred = c.predict(xtest)
print(accuracy_score(ytest, ypred))
```

```python
c = GaussianNB()
c.fit(xtrain, ytrain)
ypred = c.predict(xtest)
print(accuracy_score(ytest, ypred))
```

```python
c = KNeighborsClassifier(n_neighbors=7)
c.fit(xtrain, ytrain)
ypred = c.predict(xtest)
print(accuracy_score(ytest, ypred))
```
